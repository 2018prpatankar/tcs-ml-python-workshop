```python
#https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.sort_values.html
country_series=pd.Series(['AU','CA','DE','ES','FR','GB','IT','NDF','NL','PT','US','other'],index=[i for i in range(12)])
all_result_index=[]
all_result_values=[]
num_rows=test_grouped_array.shape[0]
#for row in range(100):  # for checking manually
for row in range(num_rows):
    row_series=pd.Series(test_grouped_array[row])
    row_series.sort_values(ascending=False,inplace=False)
    num_recommendations=0
    for index in row_series.index:
        if num_recommendations>5:# More than 5 recommendations are not permitted
            break
        if row_series[index]>cut_off:
            #print(i)
            all_result_index.append(test_grouped_df['id'][row])
            all_result_values.append(country_series[index])
            num_recommendations+=1
all_result_series=pd.Series(all_result_values,index=all_result_index)
#all_result_series
```
